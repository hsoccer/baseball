```python
necessary = df[[bool(re.findall("(((((\d|満)塁)|(\dアウト(チェンジ)?))|(ゲームセット))$)|(ホームラン)", s)) for s in df.values.flatten()]]
```

```python
necessary
```

```python
def encode_count(out, base):
    return str(out) + str(int(base["1"])) + str(int(base["2"])) + str(int(base["3"]))
```

```python
out = 0
base = dict({"1": False, "2": False, "3": False})

event_list = [] # (before_cond, after_cond)

event_list.append(encode_count(out, base))

for s in necessary.values.flatten():
    if "ホームラン" in s:
        base = dict({"1": False, "2": False, "3": False})
        after_cond = encode_count(out, base)
        event_list.append(after_cond)
        continue
    
    info = s.split()[-1]    
    #print(info)
    if "ゲームセット" in info or "チェンジ" in info:
        base = dict({"1": False, "2": False, "3": False})
        after_cond = encode_count(3, base)
        event_list.append(after_cond)
        if "チェンジ" in info:
            out = 0
            base = dict({"1": False, "2": False, "3": False})
            after_cond = encode_count(out, base)
            event_list.append(after_cond)
            
    elif "塁" in info and "アウト" in info:
        out = int(info[0])
        base_list = info[4:-1].split(",")
        if base_list == ["満"]:
            for i in ["1", "2", "3"]:
                base[i] = True
        else:
            for i in ["1", "2", "3"]:
                if i in base_list:
                    base[i] = True
                else:
                    base[i] = False
        after_cond = encode_count(out, base)
        event_list.append(after_cond)

    elif "塁" in info:
        #print(info)
        base_list = info[:-1].split(",")
        #print(base_list)
        if base_list == ["満"]:
            for i in ["1", "2", "3"]:
                base[i] = True
        else:
            for i in ["1", "2", "3"]:
                if i in base_list:
                    base[i] = True
                else:
                    base[i] = False
        #print(base)
        after_cond = encode_count(out, base)
        event_list.append(after_cond)
    
    elif "アウト" in info:
        out = int(info[0])
        after_cond = encode_count(out, base)
        event_list.append(after_cond)
    
    else:
        print(info)
        raise
```

```python
double_transfer_list = []
for i in range(1, len(event_list)):
    before, after = event_list[i-1], event_list[i]
    if before == "3000":
        continue
    double_transfer_list.append((before, after))
```

```python
triple_transfer_list = []
for i in range(2, len(event_list)):
    before, curr, after = event_list[i-2], event_list[i-1], event_list[i]
    if before == "3000" or curr == "3000":
        continue
    triple_transfer_list.append((before, curr, after))
```
