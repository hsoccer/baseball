```python
from selenium import webdriver
import re
import pandas as pd

driver = webdriver.Chrome()
```

```python
def save_text_list(lst, name):
    res = []
    for elem in lst:
        if re.match("^\d", elem.text):
            res.append(elem.text)
    if not len(res):
        return
    pd.DataFrame(res).to_csv("./game_data/"+name+".csv", encoding="cp932")
```

```python
today = "20180822"
for i in range(300):
    date = pd.to_datetime(today) - pd.Timedelta(days=i)
    date_string = str(date.year) + "{0:02d}".format(date.month) + "{0:02d}".format(date.day) + "0"
    for j in range(1, 7):
        new_date_string = date_string + str(j)
        print(new_date_string)
        url = "https://baseball.yahoo.co.jp/npb/game/{}/text".format(new_date_string)
        try:
            driver.get(url)
        except:
            continue
        l = driver.find_elements_by_tag_name("li")
        save_text_list(l, new_date_string)
```

```python
import time
```

```python
def encode_count(out, base):
    return str(out) + str(int(base["1"])) + str(int(base["2"])) + str(int(base["3"]))
```

```python
# スポナビ
def find_info(driver, name):
    res_list = []
    out = 0
    base = dict({"1": False, "2": False, "3": False})
    
    while True:
    
        result = driver.find_element_by_xpath('//*[@id="result"]').text
        if "アウト" in result:
            out = int(result.split("アウト")[0][-1]) - 1
        
        for i in ["1", "2", "3"]:
            try:
                driver.find_element_by_xpath('//*[@id="base{}"]'.format(i)).text
                base[i] = True
            except:
                base[i] = False
                
        res_list.append(encode_count(out, base))
        
        if "アウト" in result:
            out += 1
            if int(result.split("アウト")[0][-1]) == 3:
                res_list.append(encode_count(3, dict({"1": False, "2": False, "3": False})))
                out = 0
        
        driver.find_element_by_xpath('//*[@id="btn_next"]').click()
        time.sleep(1.5)
        #print(res_list)
        
        if "試合終了" in driver.find_element_by_xpath('//*[@id="liveinfo"]').text:
            break
        
    #return res_list
    pd.DataFrame(res_list, dtype="object").to_csv("./game_num_data/"+name+".csv", encoding="cp932")
```

```python
# dmenu
# ランナーは前、アウトカウントは後
def find_info(driver, name):
    res_list = []
    out = 0
    base = dict({"1": False, "2": False, "3": False})
    
    while True:
        runners = driver.find_element_by_xpath('//*[@id="liveArea"]/div/div[2]/div[3]/div[1]')
        base = dict({"1": False, "2": False, "3": False})
        for elem in runners.find_elements_by_class_name("on"):
            if "first" in elem.get_attribute("class"):
                base["1"] = True
            if "second" in elem.get_attribute("class"):
                base["2"] = True
            if "third" in elem.get_attribute("class"):
                base["3"] = True
        res_list.append(encode_count(out, base))
        #print(encode_count(out, base))
        # アウトは更新後に更新
        out = len(driver.find_element_by_xpath('//*[@id="liveArea"]/div/div[2]/div[2]/table/tbody/tr[3]').find_elements_by_class_name("on"))
        if out == 3:
            res_list.append(encode_count(3, dict({"1": False, "2": False, "3": False})))
            #print(encode_count(out, base))
            out = 0
                
        try:
            driver.find_element_by_xpath('//*[@id="liveArea"]/table/tbody/tr/td[3]/p').click()
            time.sleep(1.5)
        except:
            #　ゲームセット
            break
        
    #return res_list
    pd.DataFrame(res_list, dtype="object").to_csv("./game_num_data/"+name+".csv", encoding="cp932")
```
