```python
import pandas as pd
import re
import os
```

```python
df = None
for file in os.listdir("./game_data/")[1:]:
    _df = pd.read_csv("./game_data/"+file, encoding="cp932", index_col=0)
    try:
        df = pd.concat([df, _df])
    except:
        df = _df
df.index = list(range(len(df)))
```

```python
df
```

```python
necessary = df[[bool(re.findall("(((((\d|満)塁)|(\dアウト(チェンジ)?))|(ゲームセット))$)|(ホームラン)", s)) for s in df.values.flatten()]]
```

```python
necessary
```

```python
def encode_count(out, base):
    return str(out) + str(int(base["1"])) + str(int(base["2"])) + str(int(base["3"]))
```

```python
out = 0
base = dict({"1": False, "2": False, "3": False})

transfer_list = [] # (before_cond, after_cond)

for s in necessary.values.flatten():
    if "ホームラン" in s:
        before_cond = encode_count(out, base)
        base = dict({"1": False, "2": False, "3": False})
        after_cond = encode_count(out, base)
        transfer_list.append((before_cond, after_cond))
        continue
    
    info = s.split()[-1]    
    #print(info)
    if "ゲームセット" in info or "チェンジ" in info:
        before_cond = encode_count(out, base)
        base = dict({"1": False, "2": False, "3": False})
        after_cond = encode_count(3, base)
        transfer_list.append((before_cond, after_cond))
        out = 0
            
    elif "塁" in info and "アウト" in info:
        before_cond = encode_count(out, base)
        out = int(info[0])
        base_list = info[4:-1].split(",")
        if base_list == ["満"]:
            for i in ["1", "2", "3"]:
                base[i] = True
        else:
            for i in ["1", "2", "3"]:
                if i in base_list:
                    base[i] = True
                else:
                    base[i] = False
        after_cond = encode_count(out, base)
        transfer_list.append((before_cond, after_cond))

    elif "塁" in info:
        #print(info)
        before_cond = encode_count(out, base)
        base_list = info[:-1].split(",")
        #print(base_list)
        if base_list == ["満"]:
            for i in ["1", "2", "3"]:
                base[i] = True
        else:
            for i in ["1", "2", "3"]:
                if i in base_list:
                    base[i] = True
                else:
                    base[i] = False
        #print(base)
        after_cond = encode_count(out, base)
        transfer_list.append((before_cond, after_cond))
    
    elif "アウト" in info:
        before_cond = encode_count(out, base)
        out = int(info[0])
        after_cond = encode_count(out, base)
        transfer_list.append((before_cond, after_cond))
    
    else:
        print(info)
        raise
```

```python
triple_transfer_list = []
for i in range(2, len(transfer_list)):
    before, curr, after = transfer_list[i-2][0], transfer_list[i-1][0], transfer_list[i][0]
    if after == "0000":
        after = "3000"
    if curr == "0000":
        continue
    triple_transfer_list.append((before, curr, after))
```
